```python
def biplot(df, x=KEY, y='Man.KW', clip=True):
    p9.options.figure_size=11,8
    figdf = df if not clip else pd.concat([
        df.head(25),
        df.tail(25),
        df.sample(n=25)
    ]).drop_duplicates('word').sort_values(x,ascending=False)
    fig = p9.ggplot(figdf, p9.aes(x=x, y=y, label='word'))
    # fig += p9.geom_smooth(method='lm',alpha=.25,size=.25)
    fig+=p9.geom_vline(xintercept=0,size=.25)
    fig += p9.geom_text(size=10)
    fig += p9.theme_minimal()
    return fig
```

```python
dfx = df[df.period=='1975-2000']
dfnow2 = dfx[dfx.word.isin({'presidents','cuddled'})].append( dfx.sample(n=100) )

biplot(dfnow2, clip=False)
```
